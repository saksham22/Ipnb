```
#default size is 6*8
fig = plt.figure(figsize=(5,6))
ax = fig.add_subplot(111)

ax.set(title='My First Plot',xlabel='X-Axis',ylabel='Y-axis',xlim=(0,5),ylim=(0,10))

ax.set_title("My Second Plot")
ax.set_xlabel("X-Axis")
ax.set_ylabel('Y-Axis')
ax.set_xlim([0,5])
ax.set_ylim([0,10])

x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y, label='linear-growth')
plt.legend()
plt.show()
```

```
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(6,8))
ax = fig.add_subplot(111)

x=[1,2,8,5]
y=[1,2,4,5]

#plt.plot([10, 12, 14, 16],'yo')
ax.plot(x, y, 'g^', x, y, 'g-')
plt.show()
```

```
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111, title="My Plot")
x = [0, 1, 2, 3, 4]
y = [0, 3, 6, 9, 12]
plt.scatter(x, y, s=[20, 60], c=['r', 'g'])
plt.show()
```
