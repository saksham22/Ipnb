```
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,8))
gd = gridspec.GridSpec(2,2)
axes1 = plt.subplot(gd[0,:],title='Plot1')
axes1.set_xticks([])
axes1.set_yticks([])
axes2 = plt.subplot(gd[1,0],title='Plot2')
axes2.set_xticks([])
axes2.set_yticks([])
axes3 = plt.subplot(gd[1,1],title='Plot3')
axes3.set_xticks([])
axes3.set_yticks([])
plt.set(title='My plots')
plt.show()
```

```
#solution
import numpy as np
t=np.arange(0.0, 5.0, 0.01)
s1=np.sin(2*np.pi*t)
s2=np.sin(4*np.pi*t)
fig=plt.figure(figsize=(8,6))
axes1 = plt.subplot(2, 1, 1, title='Sin(2pix)')
axes1.plot(t,s1)
axes2 = plt.subplot(2, 1, 2, title='Sin(4pix)')
axes2.plot(t,s2)
plt.show()
```

```
#soln 2
np.random.seed(1000)
x=np.random.rand(10)
y=np.random.rand(10)
np.sqrt(x*x + y*y)
fig = plt.figure(figsize=(8,10))
axes1=plt.subplot(2,2,1,title='Scatter plot with Upper Traingle Markers')
axes2=plt.subplot(2,2,2,title='Scatter plot with Plus Markers')
axes1.scatter(x,y,c='c',marker='^',s=80)
axes2.scatter(x,y,c='c',marker='+',s=80)
xt=[0.0, 0.4, 0.8, 1.2] 
yt=[-0.2, 0.2, 0.6, 1.0]
axes1.set_xticks(xt)
axes2.set_xticks(xt)
axes1.set_yticks(yt)
axes2.set_yticks(yt)
axes3=plt.subplot(2,2,3,title='Scatter plot with Circle Marker')
axes3.scatter(x,y,c='c',marker='o',s=80)
axes3.set_xticks(xt)
axes3.set_yticks(yt)
axes4=plt.subplot(2,2,4,title='Scatter plot with Diamond Markers')
axes4.scatter(x,y,c='c',marker='d',s=80)
axes4.set_xticks(xt)
axes4.set_yticks(yt)
plt.tight_layout()
plt.show()
```

```
x=np.arange(1, 101)
y1 = x
y2 = x**2
y3 = x**3
fig = plt.figure(figsize=(8,6))
g = gridspec.GridSpec(2,2)
axes1=plt.subplot(g[0,0],title='y = x')
axes2=plt.subplot(g[1,0],title='y = x**2')
axes3=plt.subplot(g[:,1],title='y = x**3')
axes1.plot(x,y1)
axes2.plot(x,y2)
axes3.plot(x,y3)
plt.tight_layout()
plt.show()
```
